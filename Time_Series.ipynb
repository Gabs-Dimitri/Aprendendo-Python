```python
def plotar(titulo, labelx, labely, x, y, dataset):
  sns.set_palette('Accent')
  sns.set_style('darkgrid')
  ax = sns.lineplot(x = x, y = y, data = dataset)
  ax.figure.set_size_inches(12,6)
  ax.set_title(titulo, loc = 'left', fontsize = 18)
  ax.set_xlabel(labelx, fontsize = 14)
  ax.set_ylabel(labely, fontsize = 14)
  ax = ax
```

```python
plotar('Aumento das Vendas da Alucar de 2017 e 2018', 'Tempo', 'Aumento', 'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Aceleração das Vendas da Alucar', 'Tempo', 'Aceleração', 'mes', 'aceleracao', alucar)
```

```python
plt.figure(figsize=(16,12))
ax = plt.subplot(3, 1, 1)
ax.set_title('Análise de vendas da Alucar de 2017 e 2018', loc = 'left', fontsize = 18)
sns.lineplot(x = 'mes', y = 'vendas', data = alucar)
plt.subplot(3, 1, 2)
sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
plt.subplot(3, 1, 3)
sns.lineplot(x = 'mes', y = 'aceleracao', data = alucar)
ax = ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
  plt.figure(figsize=(16,12))
  ax = plt.subplot(3, 1, 1)
  ax.set_title(titulo, loc = 'left', fontsize = 18)
  sns.lineplot(x = x, y = y1, data = dataset)
  plt.subplot(3, 1, 2)
  sns.lineplot(x = x, y = y2, data = dataset)
  plt.subplot(3, 1, 3)
  sns.lineplot(x = x, y = y3, data = dataset)
  ax = ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'Análise das vendas da alucar')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
ax = plt.figure(figsize=(12, 6))
ax.suptitle('Correlação do aumento', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(alucar['aumento'][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12, 6))
ax.suptitle('Correlação da aceleração', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(alucar['aceleracao'][2:])
ax = ax
```

# Pra saber mais
## **Quando não aplicar análise de time series**
Sabemos que time series ou series temporais são uma forma de organizar no tempo as informações quantitativas e que podemos extrair muitas informações deste tipo de dado. Porém, nem sempre podemos aplicar a análise de time series.

Abaixo temos duas recomendações de quando não aplicar análise de time series:

**Não temos a variável tempo**
A principal variável em uma time series é o tempo. Sendo assim, a ausência desta variável impossibilita análise de séries temporais.

**As variáveis ou valores são constantes**
Analisando o gráfico acima de uma loja que vende 5000 caixas de chocolates há 18 anos, não podemos aplicar análises de séries temporais para extrair mais informações dos dados.

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolura['vendas'], period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({'observacao':observacao, 'tendencia':tendencia, 'sazonalidade':sazonalidade, 'ruido':ruido})
resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(resultado.index, 'observacao', 'tendencia', 'sazonalidade', resultado, 'Exemplo de Statsmodels')
```
